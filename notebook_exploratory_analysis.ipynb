```python

# exploratory_analysis.ipynb

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# 2. Load the Data
# Load the traffic data from CSV
df = pd.read_csv(url)

# Display the first few rows
df.head()
```

```python
# 3. Basic Data Information
# Show dataset info (column types, missing values, etc.)
df.info()

# Summary statistics for numerical columns
df.describe()
```

```python
# 4. Handling Missing Data
# Check for missing values
df.isnull().sum()

# If there are missing values, fill them (Example: fill missing weather data)
df['Weather Conditions'].fillna('Clear', inplace=True)
```

```python
# 5. Distribution of Traffic Flow Categories
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Traffic Flow')
plt.title('Distribution of Traffic Flow Categories')
plt.xlabel('Traffic Flow')
plt.ylabel('Count')
plt.show()
```

```python
# 6. Correlation Heatmap
# Create a heatmap to visualize correlations between numeric features
# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Traffic Data Features')
plt.show()
```

```python
# 7. Relationship between Speed and Vehicle Count
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Speed (km/h)', y='Vehicle Count', hue='Traffic Flow', palette='coolwarm')
plt.title('Speed vs Vehicle Count by Traffic Flow')
plt.xlabel('Speed (km/h)')
plt.ylabel('Vehicle Count')
plt.show()
```

```python
# 8. Traffic Flow vs Time of Day
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Hour of Day', y='Speed (km/h)', hue='Traffic Flow', palette='coolwarm')
plt.title('Traffic Flow vs Speed by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Speed (km/h)')
plt.show()
```

```python
# 9. Traffic Flow vs Weather Conditions
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Weather Conditions', y='Vehicle Count', hue='Traffic Flow', palette='Set2')
plt.title('Traffic Flow vs Vehicle Count by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Vehicle Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Rush Hour Analysis
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Hour of Day', hue='Traffic Flow', palette='coolwarm')
plt.title('Traffic Flow During Rush Hour vs Non-Rush Hour')
plt.xlabel('Rush Hour (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```
